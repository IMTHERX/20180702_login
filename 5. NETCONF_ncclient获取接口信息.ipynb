```python
from ncclient import manager
```

```python
device = {
    "host":"192.168.123.223",
    "port":22,
    "username":"lab",
    "password":"lab123",
    "hostkey_verify":False,
    "device_params":{'name':'junos'}}
```

记下来修改以上代码，通过发送获取接口信息的xml请求，处理返回后的xml信息。

```python
nc = manager.connect(**device)
```

```python
interfaces = nc.rpc("<get-interface-information><terse/> </get-interface-information>")
```

```python
routes = nc.rpc("<get-route-information><terse/> </get-route-information>")
```

```python
print(routes)
```

```python
print(interfaces)
```

```python
for interface in interfaces.xpath('//physical-interface/name'):
    print(interface.text)
```
